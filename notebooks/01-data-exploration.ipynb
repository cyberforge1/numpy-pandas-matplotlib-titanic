```python
# Import necessary libraries
import pandas as pd

# Load the dataset
# This cell loads the cleaned Titanic dataset for exploration.
file_path = '../data/processed/cleaned_titanic.csv'
df = pd.read_csv(file_path)

# Display the first few rows
print("Dataset loaded successfully.")
df.head()
```

```python
# Display basic information about the dataset
# This cell provides an overview of the dataset, including data types and non-null counts.
print("Basic Information:")
df.info()

# Summary statistics
# This cell computes and displays summary statistics for numerical and categorical columns.
print("\nSummary Statistics:")
df.describe(include='all')
```

```python
# Check for missing values
# This cell checks for missing values in each column of the dataset.
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Dynamically detect categorical columns
# This cell identifies categorical columns and displays value counts for each.
categorical_columns = df.select_dtypes(include=['bool', 'category', 'object']).columns

for col in categorical_columns:
    print(f"Value counts for {col}:")
    print(df[col].value_counts())
    print()
```

```python
# Compute and display correlation matrix
# This cell computes the correlation matrix and visualizes it using a heatmap.
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

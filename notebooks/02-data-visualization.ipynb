```python
# Import pandas for data manipulation and matplotlib for visualization
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# This cell ensures the dataset is loaded for visualization
file_path = '../data/processed/cleaned_titanic.csv'
df = pd.read_csv(file_path)
```

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt

# Plot distributions of numerical features
# This cell creates histograms for numerical features such as Age, Fare, and FamilySize
numerical_columns = ['Age', 'Fare', 'FamilySize']
df[numerical_columns].hist(bins=15, figsize=(15, 5), layout=(1, 3), edgecolor='black')
plt.suptitle("Distributions of Numerical Features")
plt.show()
```

```python
# Import pandas and matplotlib for visualization
import pandas as pd
import matplotlib.pyplot as plt

# Plot survival rates by FamilySize
# This cell analyzes survival rates based on FamilySize and visualizes the results with a bar plot
survival_by_family_size = df.groupby('FamilySize')['Survived'].mean()
print("Survival rates by FamilySize:")
print(survival_by_family_size)

survival_by_family_size.plot(kind='bar', figsize=(10, 5), title='Survival Rates by Family Size', color='skyblue')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Import seaborn and matplotlib for advanced visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot of Age vs. Fare with survival as color
# This cell creates a scatter plot of Age vs. Fare, with points colored by survival status
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm', alpha=0.6)
plt.title('Age vs. Fare with Survival Highlighted')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Import pandas for data manipulation
import pandas as pd

# Create AgeGroup feature and analyze survival rates
# This cell bins Age into groups (Child, Teenager, Adult, Senior) and analyzes survival rates for each group
bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# One-hot encode AgeGroup
age_group_dummies = pd.get_dummies(df['AgeGroup'], prefix='AgeGroup', drop_first=True)
df = pd.concat([df, age_group_dummies], axis=1)

# Survival rates by AgeGroup
age_group_columns = [col for col in df.columns if col.startswith('AgeGroup_')]

if not age_group_columns:
    print("No AgeGroup columns found in the dataset.")
else:
    for col in age_group_columns:
        survival_rate = df[df[col] == 1]['Survived'].mean()
        print(f"Survival rate for {col.replace('AgeGroup_', '')}: {survival_rate:.2f}")
```

```python
# Import seaborn and matplotlib for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to visualize Fare outliers
# This cell creates a boxplot to detect outliers in the Fare feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Fare')
plt.title('Boxplot of Fare')
plt.show()
```

```python
# Import seaborn and matplotlib for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Analyze survival rate by age and gender
# This cell visualizes survival rates based on Age and Gender using a scatter plot
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=df,
    x='Age',
    y='Survived',
    hue='Sex_male',
    palette={0: 'blue', 1: 'orange'},
    alpha=0.6
)

plt.title('Survival Rate by Age and Gender', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Survived', fontsize=12)
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Import seaborn and matplotlib for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Survival rate by Passenger Class and Gender
# This cell visualizes survival rates for combinations of Passenger Class and Gender using a bar plot
plt.figure(figsize=(12, 6))

sns.barplot(
    data=df,
    x='Pclass',
    y='Survived',
    hue='Sex_male',
    palette={0: 'blue', 1: 'orange'}
)

plt.title('Survival Rate by Passenger Class and Gender', fontsize=16)
plt.xlabel('Passenger Class (Pclass)', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
